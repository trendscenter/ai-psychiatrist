```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


```

```python

# CSV
csv_path = '/home/users/nblair7/ai-psychiatrist/merged_fixed.csv'
df = pd.read_csv(csv_path)

# Drop non-numeric or unnecessary columns
df_scores = df[['coherence', 'completeness', 'specificity', 'accuracy']]
df_scores.columns = ['Coherence', 'Completeness', 'Specificity', 'Accuracy']
# Create box plot
sns.boxplot(data=df_scores)
plt.ylim(3.0, 10.5)  # or whatever range makes sense for your data

# Customize plot
plt.title('Qualitative Evaluation Scores', fontweight = 'bold')
plt.xlabel('Metric', fontsize = 12, fontweight = 'bold')
plt.ylabel('Score', fontsize = 12, fontweight = 'bold')  
plt.yticks([3,4,5,6,7,8,9,10], fontsize = 12)
plt.xticks( fontsize = 12)
plt.grid(True)
plt.show()
plt.savefig('fig.png', dpi=1000, )  # Save the figure
```
